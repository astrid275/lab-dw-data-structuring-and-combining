In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd

# Load the dataset
url1 = "cleaned_df.csv"
df1 = pd.read_csv(url1)
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df2 = pd.read_csv(url2)
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df3 = pd.read_csv(url3)
```

```python
df1
```

```python
df1 = df1.drop('Unnamed: 0', axis=1)
```

```python
df1.columns
```

```python
df2.columns = df2.columns.str.replace(' ', '_')
df3.columns = df3.columns.str.replace(' ', '_')
```

```python
df2.columns = df2.columns.str.lower()
df3.columns = df3.columns.str.lower()
```

```python
df2.rename(columns={'st': 'state'}, inplace=True)
df3.rename(columns={'st': 'state'}, inplace=True)
```

```python
df1.head()
```

```python
df1.rename(columns= {'monthly_premium':'monthly_premium_auto', 'number_open_complaints' : 'number_of_open_complaints'}, inplace=True)
print(df1.columns)
```

```python
df3.columns
```

```python
df1.head()
```

```python
merged_df=pd.concat([df1, df2, df3], axis=0)
```

```python
merged_df
```

```python
merged_df.gender = merged_df.gender.apply(lambda x: "F" if x in ("F", "Femal", "female") else "M")
```

```python
states_mapping = {
    "Washington" : "Washington",
    "Arizona" : "Arizona",
    "Nevada" : "Nevada",
    "California" : "California",
    "Oregon" : "Oregon",
    "Cali" : "California",
    "AZ" : "Arizona",
    "WA" : "Washington"
}
merged_df.state = merged_df.state.map(states_mapping)
```

```python
merged_df.education = merged_df.education.apply(lambda x: "Bachelor" if x == "Bachelors" else x)
```

```python
merged_df.customer_lifetime_value = merged_df.customer_lifetime_value.apply(lambda x: float(str(x).strip("%")))
```

```python
luxury_mapping = {
    "Sports Car" : "Luxury",
    "Luxury SUV" : "Luxury",
    "Luxury Car" : "Luxury",
    "Four-Door Car" : "Four-Door Car",
    "Two-Door Car" : "Two-Door Car",
    "SUV" : "SUV"
}
merged_df.vehicle_class = merged_df.vehicle_class.map(luxury_mapping)
```

```python
merged_df.customer_lifetime_value = merged_df.customer_lifetime_value.astype(float)
```

```python
merged_df.number_of_open_complaints = merged_df.number_of_open_complaints.apply(lambda x: int((format(x).split("/"))[1]) if isinstance(x,str) else x)
```

```python
merged_df['number_of_open_complaints'].unique()
```

```python
merged_df.isna().sum()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_insurance_df = insurance_df.pivot_table(index='sales_channel', values='customer_lifetime_value', aggfunc='sum').round(2)
```

```python
pivot_insurance_df 
# The sales channel which has the highest customer lifetime value (including total revenue) is via Agent, 
# which is more than double the value for Web. 
# This could indicate a positive effect on sales when there is a more personal relationship with the customer.
```

```python
insurance_df['sales_channel'].unique()
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_df = insurance_df.pivot_table(index='gender', columns='education', values='customer_lifetime_value', aggfunc='mean')
```

```python
pivot_df 
```

```python
# The customer lifetime values per gender are similar through the education levels with the largest 
# difference in the high school or below column, where the Females are about 500 more than the Males. 
# There is a correlation between the higher the education the less the value, with Doctors having the 
# smallest values in both genders.
```
